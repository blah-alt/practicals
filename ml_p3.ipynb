# Gradient Descent Algorithm in Python

This notebook demonstrates the implementation of the gradient descent algorithm to find the local minima of the function $y = (x + 5)^2$ starting from $x = 3$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Defining the Function and its Derivative

We will use the following function:

$$ y = (x + 5)^2 $$

The derivative of the function is:

$$ f'(x) = 2*(x + 5) $$

```python
def function(x):
    return (x + 5) ** 2

def gradient(x):
    return 2 * (x + 5)
```

## Gradient Descent Parameters

We define the following parameters for the gradient descent:
- Learning Rate (step size)
- Initial point for x
- Precision for convergence
- Maximum number of iterations

```python
# Gradient descent parameters
cur_x = 3  # Starting value of x
learning_rate = 0.1  # Step size
precision = 0.000001  # Stopping criterion
max_iters = 10000  # Maximum iterations
prev_step_size = 1  # Large initial step size to enter the loop
x_history = [cur_x]  # To store x values
y_history = [function(cur_x)]  # To store y values (corresponding f(x))
loss_history = [function(cur_x)]  # To store loss values
```

## Gradient Descent Implementation

We will iterate using the gradient descent algorithm until convergence.

```python
# Gradient Descent Loop
iters = 0
while prev_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - learning_rate * gradient(prev_x)
    prev_step_size = abs(cur_x - prev_x)

    # Store the x, y, and loss values for plotting
    x_history.append(cur_x)
    y_history.append(function(cur_x))
    loss_history.append(function(cur_x))  # The function value is the loss

    iters += 1
```

## Visualization of Gradient Descent Path

The plot below shows the function $y = (x + 5)^2$ and the path taken by the gradient descent algorithm from the initial point to the minimum.

```python
# Create the range for the function plot
x_range = np.linspace(-14, 4, 400)
y_range = function(x_range)

# Plotting the gradient descent path
plt.figure(figsize=(14, 6))

# Function plot
plt.plot(x_range, y_range, label='y = (x + 5)^2', color='blue')
plt.scatter(x_history, y_history, color='red', zorder=5, label='GD Steps')
plt.plot(x_history, y_history, linestyle='--', color='red', label='Gradient Descent Path')
plt.title('Gradient Descent on y = (x + 5)^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```

## Visualizing the Loss Function

```python
# Loss function plot
plt.plot(range(len(loss_history)), loss_history, label='Loss Function', color='green')
plt.title('Loss Function over Iterations')
plt.xlabel('Iterations')
plt.ylabel('Loss (y = (x + 5)^2)')
plt.grid(True)
plt.legend()
```

## Final Results

The local minimum found by the gradient descent algorithm is displayed below.

```python
print(f"The local minimum occurs at x = {cur_x} after {iters} iterations.")
```
